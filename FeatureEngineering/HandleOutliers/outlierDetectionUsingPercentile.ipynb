```python
lowerLimit = df['Height'].quantile(0.01)
upperLimit = df['Height'].quantile(0.99)
```

```python
df[(df['Height']<=lowerLimit) | (df['Height']>=upperLimit)].shape
```

because I have too many outliers so trimming directly affect my data frame so I have to go for capping

# Capping

```python
df['Height'] = np.where(
    df['Height']<lowerLimit,
    lowerLimit,
    np.where(
        df['Height']>upperLimit,
        upperLimit,
        df['Height']
    )
)
```
