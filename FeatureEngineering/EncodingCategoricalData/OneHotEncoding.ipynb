## OneHotEncoding Using Pandas

### 1. One Hot Encoding on K columns
here I only transfer fuel and owner
so change the fule column replace with put 4 different column of n unique values and samely change owner column and replace with 5 different column of n unique values

```python
pd.get_dummies(df,columns=['fuel','owner'])
```

### 2. One Hot Encoding on K-1 columns
because their a problem occur if we use linear model like linear regression which called multicolinearlity problem

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```
