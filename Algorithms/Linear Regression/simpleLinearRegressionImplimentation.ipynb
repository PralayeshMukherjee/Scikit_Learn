```python
import pandas as pd
import numpy as np
```

```python
class simpleLinearRegression:
    def __init__(self):
        self.m = None
        self.b = None
    
    def fit(self,x_train,y_train):
        num = 0
        den = 0
        x_mean = np.mean(x_train)
        y_mean = np.mean(y_train)
        for i in range(x_train.shape[0]):
            xi = x_train.iloc[i]
            yi = y_train.iloc[i]
            first = xi-x_mean
            second = yi-y_mean
            num = num + first*second
        for i in range(x_train.shape[0]):
            xi = x_train.iloc[i]
            third = pow((xi-x_mean),2)
            den = den + third
        self.m = num/den
        self.b = y_mean - self.m*x_mean
    def predict(self,x_test):
        y_pred = self.m*x_test+self.b
        return y_pred
```
