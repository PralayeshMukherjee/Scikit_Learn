```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x,y = load_diabetes(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
#Linear Regression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)
print("R2 Score for Linear Regression: ",r2_score(y_test,y_pred_lr))
```

```python
#Ridge Regression
ridge = Ridge(alpha=0.1)
ridge.fit(x_train,y_train)
y_pred_ridge = ridge.predict(x_test)
print("R2 Score for Ridge Regression: ",r2_score(y_test,y_pred_ridge))
```

```python
#Lasso Regression
lasso = Lasso(alpha=0.01)
lasso.fit(x_train,y_train)
y_pred_lasso = lasso.predict(x_test)
print("R2 Score for Lasso Regression: ",r2_score(y_test,y_pred_lasso))
```

```python
#ElasticNet Regression
elasticNet = ElasticNet(alpha=0.01,l1_ratio=0.9)
elasticNet.fit(x_train,y_train)
y_pred_elasticNet = elasticNet.predict(x_test)
print("R2 Score for ElasticNet Regression: ",r2_score(y_test,y_pred_elasticNet))
```

here l1_ratio means 90% weightage of ridge regression and 10% of Lasso Regression
