### 1. How are coefficients affected?

```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
data = load_diabetes()
df = pd.DataFrame(data.data,columns=data.feature_names)
df['TARGET'] = data.target
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('TARGET',axis=1),df['TARGET'],test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import Lasso
```

```python
coeffs = []
r2_scores = []
for i in (0,0.1,1,10):
    lasso = Lasso(alpha=i)
    lasso.fit(x_train,y_train)
    coeffs.append(lasso.coef_.tolist())
    y_pred = lasso.predict(x_test)
    r2_scores.append(r2_score(y_test,y_pred))
```

```python
print(coeffs)
print(r2_scores)
```

```python
# we can see that as we increase the value of alpha, the coefficients are getting smaller and at alpha = 10, all the coefficients are zero
```

### 2. Higher Coefficients are affected more

```python
alphas = [0,0.0001,0.001,0.01,0.1,1,10,100,1000,10000]
coefs = []
for i in range(len(alphas)):
    lasso = Lasso(alpha=alphas[i])
    lasso.fit(x_train,y_train)
    coefs.append(lasso.coef_.tolist())
```

```python
arr = np.array(coefs)
coef_df = pd.DataFrame(arr,columns=data.feature_names)
coef_df['alpha'] = alphas
coef_df.set_index('alpha')
```

```python
# if we increase the lamda value then the coefficients are decreasin rapidly and it can reach to 0
```

3. Bias-Variance Trade-offIncreasing $\lambda$ (alpha): Reduces overfitting.Bias: Increases as $\lambda$ grows because the model becomes less flexible.Variance: Decreases as coefficients shrink.Goal: Pick an intermediate $\lambda$ where bias and variance reach an optimal balance point.

4. Loss Function Space & SparsityShape Distortion: Increasing $\lambda$ distorts the loss function curve unevenly rather than just smoothly shifting it.Hard Zero Point: Higher $\lambda$ pulls the minimum point of the loss function directly onto the axis where the coefficient is exactly zero.Feature Selection: Once a coefficient hits zero, further increases in $\lambda$ keep it locked precisely at zero. This is why Lasso creates true sparsity and acts as an automatic feature selector.
